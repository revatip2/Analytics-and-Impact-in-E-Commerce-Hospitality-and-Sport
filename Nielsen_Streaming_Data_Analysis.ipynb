```python
# non_english_pattern = '[^a-zA-Z0-9\s]'
hulu_data.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
hulu_data = hulu_data[~hulu_data['Title'].astype(str).str.contains(r'\s')]
has_empty_strings = hulu_data['Title'].str.strip().eq('').any()
hulu_data = hulu_data[hulu_data['Title'].str.strip().ne('')]
```

```python
streaming_data = pd.concat([netflix_data, amazon_data, hulu_data])
```

#### Streaming Data

```python
pivot_table_genre = pd.pivot_table(streaming_data, values='Title',
                                   index='Genre',
                                   aggfunc='count').sort_values(by='Title', ascending=False)
pivot_table_genre.head(10)
```

```python
top_10_genres = pivot_table_genre.head(10)

plt.figure(figsize=(10, 6))
top_10_genres.plot(kind='bar', color='skyblue')
plt.title('Top 10 Genres by Number of Titles')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
streaming_data['Duration Group'] = streaming_data['Duration Group'].str.strip()
```

```python
titles_per_year = streaming_data.groupby('YearMonth')['Genre'].count()
titles_per_year.plot(kind='bar', xlabel='Year-Month', ylabel='Genre', title='Number of Titles Released Each Year-Month')
plt.show()
```

```python
count_no_group = (streaming_data['Duration Group'] == 'No Group').sum()
# print(count_no_group)
streaming_data= streaming_data[streaming_data['Duration Group'] != 'No Group']
streaming_data['Duration Group'] = streaming_data['Duration Group'].replace('300+ minutes', '300')
```

```python
streaming_data['Avg Duration'] = streaming_data['Duration Group'].apply(lambda x: int(x.split(' ')[0]))
```

```python
filtered_data = streaming_data[streaming_data['Genre'] == 'Crime & Thriller - Drama']

average_duration = filtered_data['Duration'].mean()
streaming_data['Duration'] = streaming_data['Duration'].str.strip()
streaming_data['Duration'] = pd.to_numeric( streaming_data['Duration'], errors='coerce')
streaming_data['Duration']=streaming_data['Duration'].fillna(0)
streaming_data['Duration'] = streaming_data['Duration'].astype(int)
```

```python
average_duration = filtered_data['Duration'].mean()
print('Average Duration: ', average_duration)

content_type_counts = filtered_data['Content Type'].value_counts()
plt.figure(figsize=(4,4))
plt.pie(content_type_counts, labels=content_type_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title(f'Content Types Distribution under Crime & Thriller - Drama')
plt.show()
```

```python
average_watch_duration_by_genre = streaming_data.groupby('Genre')['Avg Duration'].mean().sort_values(ascending=False)
top_10_genres = average_watch_duration_by_genre.head(10)

plt.figure(figsize=(10, 6))
top_10_genres.plot(kind='bar', color='skyblue')
plt.title('Average Duration by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Duration')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
selected_genre = 'Crime & Thriller - Drama'
filtered_data = streaming_data[streaming_data['Genre'] == selected_genre]

average_duration_by_title = filtered_data.groupby('Title')['Avg Duration'].mean().sort_values(ascending=False)

top_10_shows = average_duration_by_title.head(10)

plt.figure(figsize=(10, 6))
plt.bar(top_10_shows.index, top_10_shows, color='skyblue')
plt.title(f'Top 10 Unique Shows in {selected_genre} by Average Duration')
plt.xlabel('Show Title')
plt.ylabel('Average Duration')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
genre_counts = streaming_data['Genre'].value_counts().sort_values(ascending=False)

most_popular_primary_genre = genre_counts.index[0]

most_popular_primary_genre_data = streaming_data[streaming_data['Genre'] == most_popular_primary_genre]

duration_counts_within_primary_genre = most_popular_primary_genre_data.groupby('Duration Group')['Title'].count().sort_values(ascending=False)

plt.figure(figsize=(18,4))
sns.barplot(x=duration_counts_within_primary_genre.index, y=duration_counts_within_primary_genre.values, palette="viridis")
plt.title(f"Duration Distribution within the Most Popular Primary Genre: {most_popular_primary_genre}")
plt.xlabel("Duration Group")
plt.ylabel("Count")
plt.show()
```

```python
top_genres = streaming_data['Genre'].value_counts().head(5).index
top_duration_groups = {}

for genre in top_genres:
    genre_data = streaming_data[streaming_data['Genre'] == genre]
    top_duration_group = genre_data['Duration Group'].value_counts().idxmax()
    top_duration_groups[genre] = top_duration_group

top_duration_df = pd.DataFrame(list(top_duration_groups.items()), columns=['Genre', 'Top Duration Group'])
top_duration_df
```
