#### 1. True Shooting Percentage

The true shooting percentage measures a player's efficiency at shooting the ball. True shooting percentage equals half the points scored divided by the sum of the field goals attempted and 0.475 times the free throws attempted.

```python
ts_df = data[['player', 'points', 'event_type']]
shots = ts_df[ts_df['event_type'].isin(['shot', 'free throw'])]

player_stats = shots.groupby('player').apply(lambda x: pd.Series({
    'Total Points': x['points'].sum(),
    'Non Free Throw Attempts': x[x['event_type'] == 'shot']['event_type'].count(),
    'Free Throw Attempts': x[x['event_type'] == 'free throw']['event_type'].count()})).reset_index()

player_stats['TS%'] = player_stats['Total Points']/(2*(player_stats['Non Free Throw Attempts']+0.475*player_stats['Free Throw Attempts']))
```

```python
player_stats.sort_values('TS%', ascending=False).head(10)
```

```python
player_stats.sort_values('TS%', ascending=True).head(10)
```

High TS% indicates efficient scoring from the field, considering both non free throws and free throws. Players like Daryl Macon and Zach Norvell Jr. have exceptionally high TS%, suggesting they are highly efficient scorers. Conversely, players with low TS%, such as Amile Jefferson and Moses Brown, may be considered overvalued if their traditional metrics do not align with their efficiency.

Next, let's analyze the traditional metrics.

#### 2. Traditional Metrics
* Points Per Game (PPG)
* Assists Per Game (APG)
* Steals Per Game (SPG)
* Blocks Per Game (BPG)

```python
player_totals = data.groupby('player').agg({'points': 'sum','assist': lambda x: x.notnull().sum(),'block': lambda x: x.notnull().sum(),'steal': lambda x: x.notnull().sum()}).reset_index()
```

```python
player_totals.head()
```

I have the player stats now, I want to check the total games played.

```python
player_games = data.groupby('player')['game_id'].nunique().reset_index()
stats_games = pd.merge(player_totals, player_games, on='player')
stats_games.rename(columns = {'game_id':'games'}, inplace=True)
```

```python
stats_games['PPG'] = stats_games['points']/stats_games['games']
stats_games['APG'] = stats_games['assist']/stats_games['games']
stats_games['BPG'] = stats_games['block']/stats_games['games']
stats_games['SPG'] = stats_games['steal']/stats_games['games']
```

```python
stats_games.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

merged_data = pd.merge(player_stats, stats_games, on='player', how='inner')
scaler = MinMaxScaler()

columns_to_normalize = ['TS%', 'PPG', 'APG', 'BPG', 'SPG']
merged_data[columns_to_normalize] = scaler.fit_transform(merged_data[columns_to_normalize])

merged_data['Score_Combined'] = merged_data[columns_to_normalize].sum(axis=1)

merged_data_sorted = merged_data.sort_values('Score_Combined', ascending=False)
```

```python
print('Undervalued Players:')
merged_data_sorted.head(10)
```

```python
print('Overvalued Players:')
merged_data_sorted.tail(10)
```


Undervalued Players: Players like Trae Young and James Harden top the list of potentially undervalued players. Despite their high TS% and significant contributions across traditional metrics, their overall impact might not be fully recognized. This suggests they could provide more value, making them valuable for their teams.

Overvalued Players: Players such as Antonius Cleveland and Matt Mooney are identified as potentially overvalued. Their lower TS% combined with traditional metrics results in a lower "Score_Combined," suggesting their perceived value might be higher than their actual impact.

This analysis provides a nuanced view of player valuation, combining efficiency and traditional statistical contributions to highlight discrepancies between perceived and actual value.

From the initial analysis of TS%, it is observed that Daryl Macon and Zach Norvell Jr. have a high TS%, and players like Amile Jefferson and Moses Brown have a low TS%. I want to now analyze how their traditional metrics are and if they are potentially undervalued or overvalued.

```python
specific_players = ['Daryl Macon', 'Zach Norvell Jr.', 'Amile Jefferson', 'Moses Brown']
specific_players_data = merged_data_sorted[merged_data_sorted['player'].isin(specific_players)]
specific_players_data[['player', 'TS%', 'PPG', 'APG', 'BPG', 'SPG', 'Score_Combined']]
```

The data for Daryl Macon and Zach Norvell Jr. shows they have exceptionally high True Shooting Percentages (TS%), suggesting they are highly efficient scorers. However, their traditional metrics are relatively low, which might indicate they are undervalued based on traditional metrics alone.

Conversely, Amile Jefferson and Moses Brown have lower TS%, indicating less scoring efficiency. Their traditional metrics are also on the lower side, but not significantly different from those of Daryl Macon and Zach Norvell Jr., suggesting that their value might be more accurately reflected by traditional metrics.

This analysis suggests that players with high TS% but low traditional metrics, like Daryl Macon and Zach Norvell Jr could be undervalued assets, while those with low TS% and traditional metrics, like Amile Jefferson and Moses Brown, may not necessarily be overvalued but rather accurately valued  depending on other factors not captured by these metrics.

#### 3. Final quarter performance analysis

Next, I analyzed the top players performing in the last few minutes of the game. This analysis helps with understanding who does well in critical situations.

```python
player_performance = data[(data['period'] >= 4) & (data['remaining_time'] != '00:00:00')].groupby('player').agg({'points': 'sum', 'play_id': 'count'}).reset_index()
player_performance = player_performance.sort_values(by='points', ascending=False)
player_performance.head(10)
```

The table above lists the top 10 NBA players who performed exceptionally well in late-game situations (4th quarter or overtime) during the 2019-2020 season, based on the total points scored. LeBron James leads the list with 644 points across 1129 plays, followed by James Harden with 599 points in 916 plays, and Donovan Mitchell with 536 points in 806 plays. This analysis highlights players who excel under pressure and contribute significantly to their teams' success in critical moments of the game.

```python
overall_performance = data.groupby('player').agg({'points': 'sum', 'play_id': 'count'}).reset_index()

merged_performance = player_performance.merge(overall_performance, on='player', suffixes=('_late_game', '_overall'))
merged_performance['late_game_points_%'] = (merged_performance['points_late_game'] / merged_performance['points_overall']) * 100
merged_performance = merged_performance.sort_values(by='late_game_points_%', ascending=False)

merged_performance.head(10)
```

The comparison reveals players whose late-game points accounted for 100% of their overall season points, indicating that all their scoring contributions occurred during late-game situations. This list includes players like Tyler Cook, Kyle Guy, and Justin Patton, among others. It's important to note that these players may have had limited playing time, which is why their late-game contributions represent such a significant portion of their overall performance. This analysis highlights the impact of players in critical moments relative to their total season contributions. Players with a record of performing well under pressure can be prioritized during critical moments of the game.
